```python
S = np.load("/media/hdd01/sklee/cont_shglm/gumbelS_test_V.npy")
NS = np.load("/media/hdd01/sklee/cont_shglm/gumbelNS_test_V.npy")

ref = S[0]
S = S[1]
NS = NS[1]
```

```python
wind_size = 250
score_array = np.empty((3 , S.shape[0] - wind_size))

for i in tnrange(score_array.shape[1]):
    score_array[0,i] = i+wind_size//2
    score_array[1,i] = metrics.explained_variance_score(y_true=ref[i:i+wind_size],
                                                      y_pred=S[i:i+wind_size],
                                                      multioutput='uniform_average')
    score_array[2,i] = metrics.explained_variance_score(y_true=ref[i:i+wind_size],
                                                      y_pred=NS[i:i+wind_size],
                                                      multioutput='uniform_average')
```

```python
fig, axs = plt.subplots(nrows=2, figsize = (16,8))
axs[1].plot(score_array[0,:5000-wind_size], score_array[2,:5000-wind_size], c="blue", label="No-Spike", linewidth=1)
axs[1].plot(score_array[0,:5000-wind_size], score_array[1,:5000-wind_size], c="red", label="Spike", linewidth=1)
axs[0].plot(ref[:5000]+14, c="black", linewidth=1)
axs[0].plot(NS[:5000], c="blue", label="No-Spike: 0.93914", linewidth=1)
axs[0].plot(S[:5000]-10, c="red", label="Spike: 0.94874", linewidth=1)

axs[1].legend()
axs[0].legend()
axs[1].set_xlabel("Time (ms)")
axs[0].set_ylabel("mV")
axs[1].set_ylabel("Variance explained /n (250 ms window)")


fig.suptitle("L2/3 (Low Burst?) shGLM vs. hGLM (5 subunits)", fontsize=16)

plt.savefig("/media/hdd01/sklee/cont_shglm/L23_sub5_low_burst.png", dpi=150, bbox_inches='tight')
plt.close()
```
