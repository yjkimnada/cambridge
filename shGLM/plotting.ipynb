```python
import numpy as np
import matplotlib.pyplot as plt
```

# Plot Syn Kernels

```python
tau = np.asarray([2.2439, 2.5602])
K = np.asarray([[[-0.0331,  0.0536],
         [ 0.0777, -0.0609]]])
K = np.mean(K[:], axis=0)
#Delta = np.asarray([4.8321])


spike_tau = np.asarray([1.9835, 2.2694])
spike_K = np.asarray([[[-0.0411,  0.0562],
         [ 0.0835, -0.0633]]])
spike_K = np.mean(spike_K[:], axis=0)


#sub_no = K.shape[0]
basis_no = K.shape[0]
max_T = 200
```

```python
t = np.arange(max_T)
t[t<0] = 0

e_kern = np.zeros((max_T))
i_kern = np.zeros((max_T))
e_spike_kern = np.zeros((max_T))
i_spike_kern = np.zeros((max_T))


for b in range(basis_no):
    t_tau = t / np.exp(tau[b])
    print(np.exp(tau[b]))
    e_kern += K[b,0] * (t_tau * np.exp(-t_tau))
    i_kern += K[b,1] * (t_tau * np.exp(-t_tau))

for b in range(basis_no):
    spike_t_tau = t / np.exp(spike_tau[b])
    print(np.exp(spike_tau[b]))
    e_spike_kern += spike_K[b,0] * (spike_t_tau * np.exp(-spike_t_tau))
    i_spike_kern += spike_K[b,1] * (spike_t_tau * np.exp(-spike_t_tau))
    
```

```python
fig, axs = plt.subplots(ncols=2, figsize = (14,7))
axs[0].plot(e_kern, label="No-Spike (Tau=9.43, 12.94)")
axs[1].plot(i_kern, label="No-Spike (Tau=9.43, 12.94)")
axs[0].plot(e_spike_kern, label="Spike (Tau=7.27, 9.67)")
axs[1].plot(i_spike_kern, label="Spike (Tau=7.27, 9.67)")
axs[0].legend()
axs[1].legend()
axs[0].set_title("Synaptic E Kernel")
axs[1].set_title("Synaptic I Kernel")
axs[0].set_ylabel("Time (ms)")
axs[1].set_ylabel("Time (ms)")
```

# Plot Spike Hist Kernels

```python
tau = np.asarray([2.0203, 2.1353, 2.1152])
K = np.asarray([[0.2513, 0.3018, 0.5003],
        [0.7128, 0.8963, 0.8831],
        [0.6878, 0.8034, 1.0254],
        [0.5086, 0.7593, 0.9318],
        [0.7588, 0.8591, 0.6981]])
K = np.mean(K, axis=0)
basis_no = K.shape[0]
max_T = 200
```

```python
t = np.arange(max_T)
t[t<0] = 0

spike_kern = np.zeros((max_T))
spike_kern = np.zeros((max_T))

for b in range(basis_no):
    t_tau = t / np.exp(tau[b])
    print(np.exp(tau[b]))
    spike_kern += K[b] * (t_tau * np.exp(-t_tau))
```

```python
plt.figure(figsize = (7,7))
plt.plot(spike_kern, label = "Tau=7.54, 8.46, 8.29")
plt.legend()
plt.xlabel("Time (ms)")
plt.title("1-Subunit shGLM (2 Synaptic Kernel) \n Spike History Kernel")
```
